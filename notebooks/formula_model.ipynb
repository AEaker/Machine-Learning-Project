```python
def find_folders(folder_name):
    directory = []
    # Grab all folder names in driectory: 
    img_directory = os.listdir(folder_name)
    for folder in img_directory:
        directory.append(os.path.join(path, folder))
    return directory
```

```python
def get_images(directory): 
    img_files = []
    for filename in os.listdir(directory):
            if filename.endswith(".jpg") or filename.endswith(".png"):
                 img_files.append(os.path.join(directory, filename))
            else:
                continue
    return img_files
```

```python
def track_faces(folders_with_image):    
    if len(sys.argv) != 3:
        print(
            "Give the path to the trained shape predictor model as the first "
            "argument and then the directory containing the facial images.\n"
            "For example, if you are in the python_examples folder then "
            "execute this program by running:\n"
            "    ./face_landmark_detection.py shape_predictor_68_face_landmarks.dat ../examples/faces\n"
            "You can download a trained facial shape predictor from:\n"
            "    http://dlib.net/files/shape_predictor_68_face_landmarks.dat.bz2")
        exit()

    predictor_path = "shape_predictor_68_face_landmarks.dat"
    faces_folder_path = folders_with_image
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor(predictor_path)
    win = dlib.image_window()
    face_aligner = openface.AlignDlib(predictor_path)
    n=0 
    
    for f in glob.glob(os.path.join(faces_folder_path, "*.jpg")):
        print("Processing file: {}".format(f))
        img = cv2.imread(f)
        win.clear_overlay()
        win.set_image(img)

        # Ask the detector to find the bounding boxes of each face. The 1 in the
        # second argument indicates that we should upsample the image 1 time. This
        # will make everything bigger and allow us to detect more faces.
        dets = detector(img, 1)
        time.sleep(2)
        print("Number of faces detected: {}".format(len(dets)), f)
        for k, d in enumerate(dets):
            print("Detection {}: Left: {} Top: {} Right: {} Bottom: {}".format(
                k, d.left(), d.top(), d.right(), d.bottom()))
            # Get the landmarks/parts for the face in box d.
            shape = predictor(img, d)
            print("Part 0: {}, Part 1: {} ...".format(shape.part(0),
                                                      shape.part(1)))
            # Draw the face landmarks on the screen.
            win.add_overlay(shape)
            time.sleep(2)

            win.add_overlay(dets)
            dlib.hit_enter_to_continue()
            # Use openface to calculate and perform the face alignment
            alignedFace = face_aligner.align(534, img, d, landmarkIndices=openface.AlignDlib.OUTER_EYES_AND_NOSE)
            time.sleep(2)
            n += 1
            dirname = 'C:/Users/Donley/Documents/GA_TECH/SUBMISSIONS/ML-Project/notebooks/detected'  + '/' + faces_folder_path.split('\\')[-1] 
            if not os.path.exists(dirname):
                os.mkdir(dirname)

            if not cv2.imwrite(os.path.join(os.path.expanduser('~'), dirname, 'aligned_{}_{}.jpg'.format(faces_folder_path.split('\\')[-1], n)), alignedFace):
                 raise Exception("Could not write image")
        
    return 
            

            # Save the aligned image to a file
#             cv2.imwrite("aligned_face_{}.jpg".format(i), alignedFace

#         # Ask the detector to find the bounding boxes of each face. The 1 in the
#         # second argument indicates that we should upsample the image 1 time. This
#         # will make everything bigger and allow us to detect more faces.
#         dets = detector(img, 1)
#         print("Number of faces detected: {}".format(len(dets)))
#         for k, d in enumerate(dets):
#             print("Detection {}: Left: {} Top: {} Right: {} Bottom: {}".format(
#                 k, d.left(), d.top(), d.right(), d.bottom()))
#             # Get the landmarks/parts for the face in box d.
#             shape = predictor(img, d)
#             print("Part 0: {}, Part 1: {} ...".format(shape.part(0),
#                                                       shape.part(1)))
#             # Draw the face landmarks on the screen.
#             win.add_overlay(shape)

#         return win.add_overlay(dets), dlib.hit_enter_to_continue()
    
```
